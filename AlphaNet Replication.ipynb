```python
#数据图片区间：20110318-20200524
#temp: 个股全部量价数据
def data_image(temp):   
    d_X = [] #数据图片
    d_Y = [] #标签:5日收益率/每日收益率
    for i,row in temp.iterrows():
        if i<29:
            continue
    
        if len(temp['open']) - i < 6:
            break
        
        matrix = np.zeros(shape=(9,30))
        matrix[0] = np.array(temp['open'][i-29:i+1])
        matrix[1] = np.array(temp['high'][i-29:i+1])
        matrix[2] = np.array(temp['low'][i-29:i+1])
        matrix[3] = np.array(temp['close'][i-29:i+1])
        matrix[4] = np.array(temp['vwap'][i-29:i+1])
        matrix[5] = np.array(temp['vol'][i-29:i+1])
        matrix[6] = np.array(temp['pct_chg'][i-29:i+1]/100)
        matrix[7] = np.array(temp['turnover_rate'][i-29:i+1])
        matrix[8] = np.array(temp['turnover_rate_f'][i-29:i+1])
    
        d_X.append(matrix)
    
        #r_t_5 = 1
        #for j in range(1,6):
            #r_t_5 *= (1+temp['pct_chg'][i+j]/100)
        
        #d_Y.append(r_t_5-1)
        d_Y.append(temp['pct_chg'][i+1]/100)
        
    return np.array(d_X), np.array(d_Y)
```

### 特征提取层

```python
#ts_corr(X,Y,d), d = 10, stride = 10
def ts_corr_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            for j in range(i+1, len(x)):
                 result.append(np.asarray([np.cov(x[i][:10],x[j][:10])[0][1]/(np.std(x[i][:10])*np.std(x[j][:10])), np.cov(x[i][10:20],x[j][10:20])[0][1]/(np.std(x[i][10:20])*np.std(x[j][10:20])), np.cov(x[i][20:30],x[j][20:30])[0][1]/(np.std(x[i][20:30])*np.std(x[j][20:30]))]))
        results.append(np.asarray(result))
           
    return np.asarray(results)            
```

```python
#ts_cov(X,Y,d), d = 10, stride = 10
def ts_cov_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            for j in range(i+1, len(x)):
                 result.append(np.asarray([np.cov(x[i][:10],x[j][:10])[0][1], np.cov(x[i][10:20],x[j][10:20])[0][1], np.cov(x[i][20:30],x[j][20:30])[0][1]]))
        results.append(np.asarray(result))
        
    return np.asarray(results)       
```

```python
#ts_stddev(X,d), d = 10, stride = 10
def ts_stddev_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            result.append(np.asarray([np.std(x[i][:10]), np.std(x[i][10:20]), np.std(x[i][20:30])]))
        results.append(np.asarray(result))
    
    return np.asarray(results)       
```

```python
def ts_stddev_10(X:9*30):
    results = np.zeros([len(X), 9, 3])
    for i in range(len(X)):
        for j in range(9):
            results[i][j] = np.array([np.std(X[i][j][:10]), np.std(X[i][j][10:20]), np.std(X[i][j][20:30])])
            
    return results
```

```python
#ts_stddev(X,d), d = 10, stride = 10
def ts_zscore_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            result.append(np.asarray([np.mean(x[i][:10])/np.std(x[i][:10]), np.mean(x[i][10:20])/np.std(x[i][10:20]), np.mean(x[i][20:30])/np.std(x[i][20:30])]))
        results.append(np.asarray(result))
        
    return np.asarray(results)    
```

```python
#ts_return(X,d), d = 10, stride = 10
def ts_return_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            
            result.append(np.asarray([(x[i][:10][-1]-x[i][:10][0])/x[i][:10][0]-1, (x[i][10:20][-1]-x[i][10:20][0])/x[i][10:20][0]-1, (x[i][20:30][-1]-x[i][20:30][0])/x[i][20:30][0]-1]))
        results.append(np.asarray(result))
        
    return np.asarray(results)
```

```python
#ts_decaylinear(X,d), d = 10, stride = 10
def ts_decaylinear_10(X:9*30):
    results = []
    sum_w = sum(range(1,11))
    for x in X:
        result = []
        for i in range(len(x)):
            weighted_means = [sum([w/sum_w*x[i][10*j+w-1] for w in range(1,11)]) for j in range(3)]
            result.append(np.asarray(weighted_means))
            
        results.append(np.asarray(result))
        
    return np.asarray(results)    
```

```python
#ts_min(X,d), d = 10, stride = 10
def ts_min_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            result.append(np.asarray([min(x[i][10*j:10*(j+1)]) for j in range(3)]))
        results.append(np.asarray(result))
        
    return np.asarray(results) 
```

```python
#ts_max(X,d), d = 10, stride = 10
def ts_max_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            result.append(np.asarray([max(x[i][10*j:10*(j+1)]) for j in range(3)]))
        results.append(np.asarray(result))
        
    return np.asarray(results) 
```

```python
#ts_sum(X,d), d = 10, stride = 10
def ts_sum_10(X:9*30):
    results = []
    for x in X:
        result = []
        for i in range(len(x)):
            result.append(np.asarray([sum(x[i][10*j:10*(j+1)]) for j in range(3)]))
        results.append(np.asarray(result))
        
    return np.asarray(results) 
```

```python
def feature_extraction_layer(X:9*30):
    return [ts_corr_10(X), ts_cov_10(X), ts_stddev_10(X), ts_zscore_10(X), ts_min_10(X), ts_decaylinear_10(X), ts_sum_10(X)]
```

### 搭建网络

```python
from keras import layers
from keras import backend as K
from keras.layers import Dense, BatchNormalization, MaxPooling2D, AveragePooling2D, concatenate, Flatten, Dropout
from keras import Input, Model
```

```python
class MinPooling2D(layers.MaxPooling2D):
    def __init__(self, pool_size=(1, 3), strides=(1,3), 
               padding='valid', data_format=None, **kwargs):
        super(MaxPooling2D, self).__init__(pool_size, strides, padding,
                                       data_format, **kwargs)

    def pooling_function(inputs, pool_size, strides, padding, data_format):
        return -K.pool2d(-inputs, pool_size, strides, padding, data_format,
                                                         pool_mode='max')
```

```python
def alphaNet():
    input1 = Input(shape = (36, 3, 1), name = 'input1')
    input2 = Input(shape = (36, 3, 1), name = 'input2')
    input3 = Input(shape = (9, 3, 1), name = 'input3')
    input4 = Input(shape = (9, 3, 1), name = 'input4')
    input5 = Input(shape = (9, 3, 1), name = 'input5')
    input6 = Input(shape = (9, 3, 1), name = 'input6')
    input7 = Input(shape = (9, 3, 1), name = 'input7')
    #Raw output 
    x1 = BatchNormalization()(input1)
    x2 = BatchNormalization()(input2)
    x3 = BatchNormalization()(input3)
    x4 = BatchNormalization()(input4)
    x5 = BatchNormalization()(input5)
    x6 = BatchNormalization()(input6)
    x7 = BatchNormalization()(input7)
    #Average Pooling
    x1_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x1)
    x2_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x2)
    x3_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x3)
    x4_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x4)
    x5_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x5)
    x6_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x6)
    x7_mean = AveragePooling2D(pool_size = (1,3), strides = (1,3))(x7)
    #Max Pooling
    x1_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x1)
    x2_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x2)
    x3_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x3)
    x4_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x4)
    x5_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x5)
    x6_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x6)
    x7_max = MaxPooling2D(pool_size = (1,3), strides = (1,3))(x7)
    #Min Pooling
    x1_min = -K.pool2d(-x1, pool_size=(1, 3), strides=(1, 3))
    x2_min = -K.pool2d(-x2, pool_size=(1, 3), strides=(1, 3))
    x3_min = -K.pool2d(-x3, pool_size=(1, 3), strides=(1, 3))
    x4_min = -K.pool2d(-x4, pool_size=(1, 3), strides=(1, 3))
    x5_min = -K.pool2d(-x5, pool_size=(1, 3), strides=(1, 3))
    x6_min = -K.pool2d(-x6, pool_size=(1, 3), strides=(1, 3))
    x7_min = -K.pool2d(-x7, pool_size=(1, 3), strides=(1, 3))
    
    #Flatten all the information
    #Raw output 
    x1 = Flatten()(x1)
    x2 = Flatten()(x2)
    x3 = Flatten()(x3)
    x4 = Flatten()(x4)
    x5 = Flatten()(x5)
    x6 = Flatten()(x6)
    x7 = Flatten()(x7)
    #Average Pooling
    x1_mean = Flatten()(BatchNormalization()(x1_mean))
    x2_mean = Flatten()(BatchNormalization()(x2_mean))
    x3_mean = Flatten()(BatchNormalization()(x3_mean))
    x4_mean = Flatten()(BatchNormalization()(x4_mean))
    x5_mean = Flatten()(BatchNormalization()(x5_mean))
    x6_mean = Flatten()(BatchNormalization()(x6_mean))
    x7_mean = Flatten()(BatchNormalization()(x7_mean))
    #Max Pooling
    x1_max = Flatten()(BatchNormalization()(x1_max))
    x2_max = Flatten()(BatchNormalization()(x2_max))
    x3_max = Flatten()(BatchNormalization()(x3_max))
    x4_max = Flatten()(BatchNormalization()(x4_max))
    x5_max = Flatten()(BatchNormalization()(x5_max))
    x6_max = Flatten()(BatchNormalization()(x6_max))
    x7_max = Flatten()(BatchNormalization()(x7_max))
    #Min Pooling
    x1_min = Flatten()(BatchNormalization()(x1_min))
    x2_min = Flatten()(BatchNormalization()(x2_min))
    x3_min = Flatten()(BatchNormalization()(x3_min))
    x4_min = Flatten()(BatchNormalization()(x4_min))
    x5_min = Flatten()(BatchNormalization()(x5_min))
    x6_min = Flatten()(BatchNormalization()(x6_min))
    x7_min = Flatten()(BatchNormalization()(x7_min))
    
    x = concatenate([x1,x2,x3,x4,x5,x6,x7,x1_mean,x2_mean,x3_mean,x4_mean,x5_mean,x6_mean,x7_mean,x1_max,x2_max,x3_max,x4_max,x5_max,x6_max,x7_max,x1_min,x2_min,x3_min,x4_min,x5_min,x6_min,x7_min])
    x = Dense(30, activation = 'relu', kernel_initializer="truncated_normal")(x)
    #x = Dropout(0.5)(x)
    output = Dense(1, activation = 'linear', kernel_initializer="truncated_normal", name = 'output')(x)
    
    model = Model(inputs = [input1,input2,input3,input4,input5,input6,input7], outputs = [output])
    #model.summary()
    
    return model
```
